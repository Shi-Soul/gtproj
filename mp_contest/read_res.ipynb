```python
import numpy as np
def split_file_by_empty_lines(file_path,skip_lines=4):
    with open(file_path, 'r') as f:
        lines = f.readlines()[skip_lines:]

    segments = []
    current_segment = []

    for line in lines:
        line = line.strip()
        if line == '':
            if current_segment:
                segments.append(''.join(current_segment))
                current_segment = []
        else:
            current_segment.append(line + '\n')

    if current_segment:
        segments.append(''.join(current_segment))

    return segments

def extract_column(segment):
    first_line = segment.split('\n')[0]
    columns = first_line.split(' ')
    columns = [column.replace('\\', '') for column in columns]
    columns = [column for column in columns if column != '']
    # print(segment)
    
    # for each line except the first one, split it according to '  ' (double space)
    # remove empty string, get each column content
    res = {col:[] for col in columns}
    for line in segment.split('\n')[1:]:
        columns_content = line.split('  ')
        columns_content = [content for content in columns_content if content != '']
        # print(columns_content)
        for i, content in enumerate(columns_content):
            if i>0:
                try:
                    var_cont = eval(content)
                except:
                    var_cont = content
                res[columns[i-1]].append(var_cont)
    return res

def format_file(file_path,skip_lines=4):
    segments = split_file_by_empty_lines(file_path,skip_lines)
    cont = {}
    for i, segment in enumerate(segments):
        # print(f"Segment {i+1}:")
        # print(extract_column(segment))
        # print()
        cont.update(extract_column(segment))
    return cont
```
